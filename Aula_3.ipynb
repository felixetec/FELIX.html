Aula de Progrmação e Algorítimos
* Calcular a área de uma sala
# Entrada

Medir o comprimento da sala
Medir a largura da sala

# Processamento

Fazer a multiplicação do comprimento pela largura

# Saída

Mostrar o resultado


```python
# Entrada

comprimento = 7 # metros
largura = 4 # metros
```

```python
# Processamento

área = comprimento * largura
```

```python
print (f" área total da sala é: {área} metros quadrados" )

```

```python
# Calcular o quadrado

#Entrada

lado = 2 # metros

#Processamento

resultado = lado * lado

#Saída

print (f" o resultado do quadrado é: {resultado}")
```
