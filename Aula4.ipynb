# Entrada : graus celsius

#Processamento
F = C * (9/5) + 32

#Saida

mostrar graus Celsius em Fahrenheit

```python
# Escrever um programa que recebe uma quantidade de graus Celsius e converter para Fahrenheit

#Entrada

c = float(input("digite a temperatura em graus celsius: "))

#Processamento

fahrenheit = c * (9 / 5) + 32

#Saída

print(f"A temperatura de firenheit é igual a: {fahrenheit}")
```
